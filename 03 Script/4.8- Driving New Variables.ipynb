# 1a. Create "price_label" column

```python
# Create a subset of first million rows

df = df_ords_prods_merge[:1000000]
```

```python
# Define a function for price labels

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Utilize the price range function. Remind yourself: Although Python can technically execute this, it is recommended to use the loc() method instead.

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Find most expensive product.

df['prices'].max()
```

```python
# Employ the loc() method on a subset of the DataFrame to implement the price range function.

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

# 1b. Create "busiest_day" column

```python
# Change "order_dow" column name to "order_day_of_week."

df_ords_prods_merge.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)
```

```python
# Find frequency of orders_day_of_week.

df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Utilize a for-loop to assess whether orders fall on the "busiest day" (0 = Saturday), "least busy" (4 = Wednesday), or "regularly busy" (other days of the week).
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create new column in ords_prods_merge to see results in context.

df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

# “Busiest day” to become “Busiest days” (plural) and identifying slowest days of the week.

```python
# Generate a new column named "busiest_days" based on the values in the "orders_day_of_week" column to categorize days as the two busiest days (0, 1), the slowest days (4, 3), or regular busy days 


result_2 = ["Busiest days" if value in [0, 1] else "Slowest days" if value in [3, 4] else "Regularly busy" for value in df_ords_prods_merge["orders_day_of_week"]]
```

# 4. Add a new column called "busiest_period_of_day" to classify time periods into categories such as "Most orders," "Average orders," and "Fewest orders." 





```python
# value counts in "order_hour_of_day" column.

df_ords_prods_merge['order_hour_of_day'].value_counts()
```

The counts of values mentioned earlier will be divided into three equal parts, and the resulting labels for these parts are as follows: "Most orders" will include values 10, 11, 14, 15, 13, 12, 16, and 9; "Fewest orders" will include values 23, 6, 0, 1, 5, 2, 4, and 3; and "Average orders" will encompass all remaining values in the order_hour_of_day column.


```python
result_3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
df_ords_prods_merge['busiest_period_of_day'] = result_3
```
